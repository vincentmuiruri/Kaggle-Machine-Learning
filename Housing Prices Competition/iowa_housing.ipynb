```python
# Importing the libraries
import pandas as pd
```

```python
# file path and loading the data
filepath = r"D:\Practice Projects\Python\Machine Learning\Kaggle Machine Learning\Housing Prices Competition\train.csv"

iowa_data = pd.read_csv(filepath)

iowa_data.head()
```

```python
# showing the features in the dataset
iowa_data.columns
```

```python
# # dropping missing values
# print("Dropping Missing values....")
# iowa_data = iowa_data.dropna(axis=0)
# print("Missing values dropped!")
```

```python
# using dot notation to assign the y variable (target variable)
y = iowa_data.SalePrice
```

```python
# choosing the features to be inputted in the model
iowa_features = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']

# assigning the features of the model to X
X = iowa_data[iowa_features]
```

```python
# show the x variable
X.head()
```

```python
# statistical summary of the features
X.shape
```

```python
# importing the sklearn library
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
dt_model = DecisionTreeRegressor(random_state=1)

# Fit model
dt_model.fit(X, y)
```

```python
# path to file you will use for predictions
test_data_path = r'D:\Practice Projects\Python\Machine Learning\Kaggle Machine Learning\Housing Prices Competition\test.csv'

# read test data file using pandas
test_data = pd.read_csv(test_data_path)

# create test_X which comes from test_data but includes only the columns you used for prediction.
# The list of columns is stored in a variable called features
test_X = test_data[iowa_features]

# make predictions which we will submit. 
test_preds = dt_model.predict(test_X)
```

```python
# after fitting the model, we use it to make predictions
print("Making predictions for the following 5 houses:")
print(test_X.head())
print("The predictions are")
print(dt_model.predict(test_X.head()))
```

```python
# Run the code to save predictions in the format used for competition scoring

output = pd.DataFrame({'Id': test_data.Id,
                       'SalePrice': test_preds})
output.to_csv('iowa_housing_submission.csv', index=False)
```

```python
# model validation 
# importing the library for MAE
from sklearn.metrics import mean_absolute_error

# assigning predicted values
predicted_home_prices = dt_model.predict(X)

# computing MAE for the model
mean_absolute_error(y, predicted_home_prices)
```

```python
# random forest model
from sklearn.ensemble import RandomForestRegressor

# Define a random forest model
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X, y)
rf_val_predictions = rf_model.predict(test_X)

# after fitting the model, we use it to make predictions
print("Making predictions for the following 5 houses:")
print(test_X.head())
print("The predictions are")
print(rf_model.predict(test_X.head()))
```
